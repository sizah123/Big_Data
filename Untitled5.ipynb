```python
#question 1
import apache_beam as beam
from apache_beam.options.pipeline_options import PipelineOptions
import csv
from datetime import datetime


def format_date(date_str):
    try:
        # Attempt to parse the date in the format 'YYYY/MM/DD'
        return datetime.strptime(date_str, '%Y/%m/%d').strftime('%Y-%m-%d')
    except ValueError:
        # If the date format is different, you can add more cases or raise an error
        raise ValueError(f"Date format for '{date_str}' not recognized.")


# Function to process the rows
def process_row(row):
    fields = next(csv.reader([row]))
    user_id, name, gender, age, address, date_joined = fields

    # Split address and handle cases with missing parts
    address_parts = address.split(',')
    city = address_parts[0].strip() if len(address_parts) > 0 else ''
    state = address_parts[1].strip() if len(address_parts) > 1 else ''
    zip_code = address_parts[2].strip() if len(address_parts) > 2 else ''

    # Format the address correctly
    formatted_address = f"{city}, {state}, {zip_code}".strip(', ')

    # Reformat the date
    formatted_date = format_date(date_joined)

    # Change gender from abbreviation to full form
    gender = "Female" if gender == 'F' else "Male"

    return f"{user_id}; {name}; {gender}; {age}; {formatted_address}; {formatted_date}"

# Beam Pipeline for task 1
def run_pipeline():
    pipeline_options = PipelineOptions()
    with beam.Pipeline(options=pipeline_options) as p:
        input_data = (p
                      | 'Read CSV File' >> beam.io.ReadFromText('users_v (1).csv', skip_header_lines=1)
                      | 'Process Rows' >> beam.Map(process_row)
                     # | 'Write to CSV' >> beam.io.WriteToText('marketing_format.csv', file_name_suffix='.csv', shard_name_template='')
                      | 'Print Results' >> beam.Map(print))

if __name__ == '__main__':
    run_pipeline()
```

```python
import apache_beam as beam
from apache_beam.options.pipeline_options import PipelineOptions
import csv



# Function to process the rows
def process_row(row):
    try:
        # Parse the row using comma as a delimiter
        fields = next(csv.reader([row], delimiter=','))
        # Ensure that there are exactly 6 fields before unpacking
        if len(fields) != 6:
            raise ValueError(f"Expected 6 fields, but got {len(fields)}: {fields}")

        user_id, name, gender, age, address, date_joined = fields
        # Split address by '-' and extract the state (second part)
        address_parts = address.split('-')
        state = address_parts[1].strip() if len(address_parts) > 1 else ''
        return state
    except Exception as e:
        # Print/log the error and return None for invalid rows
        print(f"Error processing row: {row}, Error: {e}")
        return None



# Beam Pipeline for task 1
def run_pipeline():
    pipeline_options = PipelineOptions()
    with beam.Pipeline(options=pipeline_options) as p:
        input_data = (p
                      | 'Read CSV File' >> beam.io.ReadFromText('users_v (1).csv', skip_header_lines=1)
                      | 'Process Rows' >> beam.Map(process_row)
                      # Filter out None values returned by invalid rows
                      | 'Filter None' >> beam.Filter(lambda x: x is not None)
                      | 'Count State' >> beam.combiners.Count.PerElement()
                      | 'Print Results' >> beam.Map(print))



if __name__ == '__main__':
    run_pipeline()
```
